## S2.3 - Selecting a "Piece" of the Data

---

Sometimes we only want a portion of the data. 

*   We only want certain columns, either to make the data more manegable or to save space. 
*   We only want certain rows becauswe we are interested in a certain subset of data

Below, I present several techniques you can use to selecting only certian rows. 





```
# Select only certain columns #

OnlyPrice_and_Location = data[['id', 'price', 'neighbourhood']]

OnlyPrice_and_Location.head()
```

```
# To select rows, we have to specify conditions. Let's look only at data in Santa Monica #

SantaMonicaData = data.loc[data['neighbourhood'] == 'Santa Monica']

# See Output # 
SantaMonicaData.head(2)
```

```
# We can also make more complex queries, by combining several conditions. Lets look only at "Entire Homes" in Stanta Monica # 

SantaMonicaData = data.loc[data['neighbourhood'] == 'Santa Monica']
EntireHomesInSantaMonica = SantaMonicaData.loc[SantaMonicaData['room_type'] == "Entire home/apt" ]

# Or you can do it in one line #

EntireHomesInSantaMonica = data.loc[ (data['neighbourhood'] == "Santa Monica") & (data['room_type'] == "Entire home/apt")]

# See Output # 
EntireHomesInSantaMonica.head(2)
```

```
# You can also use this to find "Or" conditions, such as Santa Monica or Culver City #

SantaMonicaOrCulverCity = data.loc[ (data['neighbourhood'] == "Santa Monica") | (data['neighbourhood'] == "Culver City")]

# See Output # 
SantaMonicaOrCulverCity.head(2)
```

## S2.3 - Adding New Variables or Combining Data
--- 

You may want to add new variables to a python data frame. For example, you might want to calculcate how far a rental is above or below the average price. 

```
# We can calculate the average price using the fucntions from above # 

data['price'].mean()
```

```
# We can add this to a new column simply as ..

data['average_price'] = data['price'].mean()
```

```
# We can then calculcate the difference as ...

data['price_gap'] = data['price'] - data['average_price']

# or all in one line with ... 

data['price_gap'] = data['price'].mean() - data['average_price']
```

Sometimes, we may want to compare the average within groups. For example, compare the apartments in Santa Monica to the average price in Santa Monica. 

Lets repeat this for Santa Monica and Culver City.

```
# Select Only Data for Santa Monica #

SM = data.loc[ (data['neighbourhood'] == "Santa Monica") & (data['room_type'] == "Entire home/apt")]

# Select Only Data for Culver City # 

Culver = data.loc[ (data['neighbourhood'] == "Culver City") & (data['room_type'] == "Entire home/apt")]

# Calculate Averages for each group # 

SM['price_gap'] = SM['price'].mean() - SM['average_price']

Culver['price_gap'] = Culver['price'].mean() - Culver['average_price']

# We get some warnings, but this is not a big deal -- Python is warning us that this is not the best form, but its clear and intuitive #
```

```
# Combining the Data - We use the Concatenate function. It adds the two datasets, end to end # 

CombinedData = pd.concat([SM, Culver])
```

## S2.3 - Merging Different Data Frames
-----
Where Python, SQL, R and other languages really start to show their use, and differentiate themselves from tools such as Excel is when it comes to merging and combining datasets.  

It is often most efficient to store datasets in smaller individual tables and then combine them as needed. (Efficient meaning it takes less space on your computer to store, and less time to load and perform operations)

```
# Load Dataset #1: Listing Data # 

listings = pd.read_csv("./LA2020listings.csv")
```

```
# Load Dataset #2: Listing Data # 

reviews = pd.read_csv("./LA2020reviews.csv")
```

```
# Let's look at both datasets to get a sense of the data # 

listings.head(5)
```

```
reviews.head(5)
```

In the *reviews* data we have a single column listing the date of each review and an identifier (listing_id). This identifier is useful becuase it also applies to the listing data table. 

Using *listing_id* we can combine the data from these different tables. 

Suppose that to the reviews data, we would like to add a column that indicates the neighbourhood of the listing and the hostname. 

```
# Step #1. Let's select only the relevant variables for the listing data # 

ColumnsFromListing = listings[['id', 'neighbourhood', 'host_name']]

ColumnsFromListing.head()
```

```
# Step #2. Merge or Join the Two datasets (Basically add the columns from one to the other)

reviews.merge(ColumnsFromListing, left_on = 'listing_id', right_on = 'id', how = 'inner')
```

Above, we can see that we now have added to the reviews data, all of the data from the listing data table. 

Let's decompose this command. 

There are two datasets: 
- The Left: Reviews. This is the dataset that is the base that we are attaching the other one to.
- The Right: ColumnsFromListing. This is the dataset we are attaching to the base dataset. 

We specify the command as : Left_Dataset.merge(Right_Dataset)

We also have to specify what to merge on. This means what are the variables that are the same across both datasets that we can use to combine. 

- left_on: indicates the name of the variable in the Left dataset. 
- right_on: indicates the name of the variable in the Right dataset. 

If the variable has the same name in both, you can only specify "on" instead of left_on, right_on. 

There are also different types of joins. 

- Inner: Keep only the values where both datasets match up (left_on and right_on is the same in both)
- Left: Keep all of the observations in the Left, but only those in the Right dataset that match up. 
- Right: Keep all of the observations in the Right, but only those in the Left dataset that match up. 
- Outer: Keep all of the observations including those that existing in both datasets, but also those that do not. 

This is indicated in the **how = 'inner'** command. If you do specify anything  then you get an inner join by default. 

You can find more details explaining these commands here. 

https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html#:~:text=merge%20is%20a%20function%20in%20the%20pandas%20namespace,,the%20index-on-index%20(by%20default)%20and%20column%20(s)-on-index%20join.

```
# We would normally save this with annother name in order to be able to access this later #

merged_data = reviews.merge(ColumnsFromListing, left_on = 'listing_id', right_on = 'id', how = 'inner')
```
