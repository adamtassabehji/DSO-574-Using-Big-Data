## S1.1 Data Structures in Python
----

For Python to use any type of data, it needs to load it from storage (a file on your hard drive) into memory (the RAM or active memory) where it can be used. You can also enter data manually, as I will do below, to use in Python. 

There are several different ways that we can define objects in Python, for them to be stored in memory and for us to be able to use it. Therefore, as a first step I will introduce the different ways that data will be defined below.

#### **S1.1.A Scalars: Integers, Floats and Strings**

---

Scalars are a simple way of assinging a single value to an object. We can assign (almost) any name we want to reference our scalars.

For instance, suppose that we want to define an Object by the name: **A**

There are three different ways that we can define this: 

*   *Integers:* Numeric values that can take on only whole numbers, e.g. 1,2,3
*   *Floats:* Numeric values that can take on decimal values, e.g. 1.0001
*   *Strings:* Text values that can take on any possible value. Must be defined within brackets. 



```python
# As an Integer # 

A = 1

A
```

```python
# As a Float # 

A = 1.00

A
```

```python
# As a String # 

A = "1"

A
```

Now, what are the benefits of saving these data as one type versus annother. 

*   Saving an object as a *Int* is more efficient than a *Float* and even more efficient than a *String*. [Strings are BY FAR the most inefficient]
*   You can perform mathematical operations with *Float* and *Int* but not string. 
*   Strings are the ONLY way to store textual data. 

#### **S1.1.B Lists and Dictionaries**

---

Scalars are often not useful and therefore we want to combine many of these into an object so that we can work with the data. The reason why will hopefully become clearer in the following section. 

Here, we will talk about two types of data structures that allow you to store many different scalars. *Lists* and *Dictionaries*.  

*   *Lists:* A colleciton of objects combined together in a single list, where you can refer to objects based on their position in a list. 

*   *Dictionary:* A collection of objects combined together in a dictioanary, where you can refer to objects based on the assigned label. 

The differences and relative benefit of each will become apparent below. We can define lists with [] and dictionaries with {}.



```python
# Define A List # 

Cities = ['New York', 'Chicago', 'Los Angeles']
States = ['NY', 'IL', 'CA']
Populations = [19216182, 	13214799, 	9458539]

```

Notice that we can define both text (string) and numeric (int or float) values within that list. 

```python
# Define A Dictionary # 

States_Dict = {'New York':'NY', 'Chicago':'IL', 'Los Angeles':'CA'}
Populations_Dict = {'New York':19216182, 'Chicago':13214799, 'Los Angeles':9458539}
```

The difference between these types of data structures is how we can refer to them.

```python
# adding [0] behind an object provides the first object in a list, [1] the first and so on. 

Cities[0]
```

```python
Cities[1]
```

```python
# We can also select a range of cities [First, Last - Stops one before]

Cities[0:2]
```

Note that counting begins at 0. So the first object is referd to as [0], the second as [1], etc. 


***In the case of dictionaries***, we can refer to an object based on it's label. 

We can also use square brackets to refer to the index [positional indicator like so]

```python
# For Example, Selecting the population that coresponds to "New York" in our dictionary

Populations_Dict['New York']
```

## S1.3 - Loops and Repetitive Functions

---

A key benefit of programming, is that you can ask a computer to perform the same operation multiple times. This sounds odd at first, but it makes more sense once you see it in action. 

The basic functions are called *loops*. There are two types of loops: 

*   For Loops: Perform a function **for** each object in the loop.
*   While Loops: Perform a function **while** a particular condition is met. 


```python
# Lets write a for loop #

for i in [1,2,3,4,5,6]:

  Result = i**2

  print(Result)
```

What is the command above doing: 

- The first line we itterate (loop or go) through each object in the list [1,2,3,4,5,6], we will perform whatever appears in the indented code below. **Note that the colon (i.e. : ) and the indentation (a single tab) are critical here!**
- In the first indented line we perform a function for each observation. Notice that we are refering to this as *i* in the indented code. We are simply dividing 10 by each value of *i*. 
- In the second indented line, we ask python to print (output to us) the resulting calculation.


Rather than specifying the list, we can also ask python to automatically generate a list with the range() function. This works as range(Start Value, End Value, Increment) 

```python
# Example of the Range Function # 

for i in range(0,100,10):
  
  print(i)
```

A less inuitive form of the loop is a while loop. Here, we have to define *i* up front, and then add a function where we itterate *i* with each run and also add a condition. The loop will run only until the while condition is satisfied.

```python
i = 1 

while i < 10: 

  i = i + 1

  print(i)
```

What is the command above doing:

- The first line we define i
- The second line we define a condition. Here, the loop will only run as long as i is less than ten. 
- Third, in the first line of the loop (i = i + 1) we increment the variable i by one each time the loop runs. This means that i will change from 1 to 2 to 3, etc as the loop runs. 
- In the final line, we print the value for i so we can see what is happening in the code. 



## S1.4 - Accesing Data in Files

---

Often, we want to use data that is stored on our computer, instead of just typing it in. Here, I will show how we can load dat ainto python. Later on, I will show more elaborate ways of doing this. 

I will work backwards here, just for the sake of illustration.  

Lets suppose that we have a dataset on the different cities. 

```python
# Lets Create a Small Dataset Manually. This is a List of Lists. 

CityData = []

CityData.append(['City', 'State', 'Population'])
CityData.append(['Los Angeles', 'CA', 9458539 ])
CityData.append(['New York', 'NY', 19216182 ])
CityData.append(['Chicago', 'IL', 13214799 ])

CityData
```

```python
# Let's save this to a data file #

File = open("./Python/Test Output.csv", 'w')


for i in CityData: 

  OutputLine = i[0] + " , " + i[1] + " , " + str(i[2]) + "\n"

  File.write(OutputLine)

File.close()
```

```python
# Lets go back and read that file #

File = open("./Python/Test Output.csv", 'r').readlines()

for i in File: 

  print(i.split(","))
```
